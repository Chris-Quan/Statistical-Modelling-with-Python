Build a regression model.

```python

import pandas as pd
import statsmodels.api as sm
import numpy as np

fullDataSet = pd.read_csv('joinedData.csv')

fullDataSet['review_count'] = fullDataSet['review_count'].replace(np.nan,0)

reviewCount = fullDataSet['review_count']
freeBikes = fullDataSet['freeBikes']

freeBikes = sm.add_constant(freeBikes)
line_reg = sm.OLS(reviewCount,freeBikes)


fullDataSet
```

Provide model output and an interpretation of the results. 

```python
model = line_reg .fit()
print_model = model.summary()
print(print_model)

#since the p value is less than 0.05 there is a strong relationship between the review count of a restaruant and the ammount of availiable bikes in the area
#I believe this is because areas with less availiable bikes are generally more busy leading to more traffic for the business in the area boosting their review count
#this could also be interpretted diffrently as well - it could be possible that because an area has less availible bikes - people are more likely to just go to whichever restaruant is closest 
#
```
