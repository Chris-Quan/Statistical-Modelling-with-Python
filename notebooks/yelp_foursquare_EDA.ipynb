# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python
bikeData = pd.read_csv('bikeData.csv')
longitude = bikeData['longitude']
latitude = bikeData['latitude']

fourSquareKey = os.environ['FOURSQUARE_API_KEY']
headers = {"accept": "application/json", "authorization": fourSquareKey}
fsqDataList = [200]

for i in range(201):

    url = 'https://api.foursquare.com/v3/places/search?ll=' + str(latitude[i]) + ',' + str(longitude[i]) +'&radius=1000&categories=13065&fields=name%2Crating%2Ccategories%2Clocation%2Cdistance'
  
    fsqData = requests.get(url,headers=headers)
    fsqDataList.append(fsqData.json())











```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
fsqDataList[i]['results'][1]['distance']
```

```python
resName = []
rating = []
resLocation = []
resDistance = []
resLatitude = []
resLongitude = []
nearbyBikeStation = []
nearbyFreeBikes =[]

bike_data = pd.read_csv('bikeData.csv')
bikeStation =  bikeData['bikeStationAddress']
freeBikes =  bikeData['freeBikes']

for i in range(1,len(fsqDataList)):
    for j in range(len(fsqDataList[i]['results'])):
        for k in range(len(fsqDataList[i]['results'][j]['categories'])):
            if 'Restaurant' in fsqDataList[i]['results'][j]['categories'][k]['name']:
                #restaurantList.append([fsqDataList[i]['results'][j],i,[i]])
                resName.append(fsqDataList[i]['results'][j]['name'])
                resLocation.append(fsqDataList[i]['results'][j]['location']['address'])
                resDistance.append(fsqDataList[i]['results'][j]['distance'])
                try:
                    rating.append(fsqDataList[i]['results'][j]['rating'])
                except KeyError:
                    rating.append(0)
                nearbyBikeStation.append(bikeStation[i])
                #nearbyFreeBikes.append(freeBikes[i])



```

Put your parsed results into a DataFrame

```python
fsqDF = pd.DataFrame(
    {
        'restaurant_name': resName,
        'address': resLocation,
        'rating' : rating,
        'bikeStationAddress' : nearbyBikeStation,
        'bikeStationDistance': resDistance,

    }
)
fsqDF.to_csv('fsqData.csv',index=False)


fsqDF
```

# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
bikeData = pd.read_csv('bikeData.csv')

latitude = bikeData['latitude']
longitude = bikeData['longitude']


yelpDataList = [200]


yelpKey = os.environ['yelp_api_key']

for i in range(201):
    urlYelp = 'https://api.yelp.com/v3/businesses/search?categories=restaurants&latitude='+ str(latitude[i])+ '&longitude='+ str(longitude[i])+ '&radius=1000'
    headers = {"accept": "application/json", "authorization": 'bearer ' + yelpKey}
    yelpData = requests.get(urlYelp, headers=headers)
    yelpDataList.append(yelpData.json())
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
#print(yelpDataList[1]['businesses'][0]['name'])
#print(yelpDataList[1]['businesses'][0]['location']['address1'])
#print(yelpDataList[1]['businesses'][0]['rating'])
#print(yelpDataList[1]['businesses'][0]['review_count'])

yelpName = []
yelpLocation = []
yelpRating = []
yelpReviewCount = []
yelpLatitude = []
yelpLongitude = []
nearbyBikeStation = []
nearbyFreeBikes =[]

bike_data = pd.read_csv('bikeData.csv')
bikeStation =  bikeData['bikeStationAddress']
freeBikes =  bikeData['freeBikes']

for i in range(1, len(yelpDataList)):
    for j in range(len(yelpDataList[i]['businesses'])):
        yelpName.append(yelpDataList[i]['businesses'][j]['name'])
        yelpLocation.append(yelpDataList[i]['businesses'][j]['location']['address1'])
        yelpRating.append(yelpDataList[i]['businesses'][j]['rating'])
        yelpReviewCount.append(yelpDataList[i]['businesses'][j]['review_count'])
        yelpLatitude.append(yelpDataList[i]['businesses'][j]['coordinates']['latitude'])
        yelpLongitude.append(yelpDataList[i]['businesses'][j]['coordinates']['longitude'])
        nearbyBikeStation.append(bikeStation[i])
        nearbyFreeBikes.append(freeBikes[i])




```

Put your parsed results into a DataFrame

```python
yelpDF = pd.DataFrame(
    {
        'restaurant_name': yelpName,
        'address': yelpLocation,
        'rating': yelpRating,
        'review_count': yelpReviewCount,
        'reslatitude': yelpLatitude,
        'reslongitude': yelpLongitude,
        'bikeStationAddress' : nearbyBikeStation,
        #'free_bikes': nearbyFreeBikes,
    }
)





yelpDF.to_csv('yelpData.csv',index=False)

yelpDF
```
